```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


base_dir = 'skin_lesion_images'
images = []
labels = []

for class_name in os.listdir(base_dir):
    class_dir = os.path.join(base_dir, class_name)
 
    if os.path.isdir(class_dir):
        image_files = [f for f in os.listdir(class_dir) if f.endswith('.jpg')]
        if image_files:
            print(f"Processing images in directory '{class_dir}'")
            for image_file in image_files:
              
                image_path = os.path.join(class_dir, image_file)
                image = cv2.imread(image_path)
                if image is not None:
                    image = cv2.resize(image, (256, 256))
                    images.append(image)
                    labels.append(class_name)
                else:
                    print(f"Error: Unable to load image '{image_file}' from directory '{class_dir}'")
        else:
            print(f"Error: No images found in directory '{class_dir}'")
    else:
        print(f"Error: '{class_dir}' is not a directory")


images_array = np.array(images)
labels_array = np.array(labels)


print("Shape of images array:", images_array.shape)
print("Shape of labels array:", labels_array.shape)
X_train, X_test, y_train, y_test = train_test_split(images_array, labels_array, test_size=0.2, random_state=42)

print("Training images shape:", X_train.shape)
print("Testing images shape:", X_test.shape)
print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.utils import to_categorical
```

```python
print("Shape of images array:", images_array.shape)
print("Shape of labels array:", labels_array.shape)
X_train, X_test, y_train, y_test = train_test_split(images_array, labels_array, test_size=0.2, random_state=42)

print("Training images shape:", X_train.shape)
print("Testing images shape:", X_test.shape)
print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)

```

```python
# Print the shape of the images array
print("Shape of images array:", images_array.shape)

# Convert the image data to float32
images_array = images_array.astype('float32')

# Normalize the pixel values to the range [0, 1]
images_array /= 255
```

```python
# Check the data type
print("Data type of images_array:", images_array.dtype)

# Check the range of pixel values
print("Minimum pixel value:", images_array.min())
print("Maximum pixel value:", images_array.max())
```

```python
# Define the mapping of class labels to numerical values
class_mapping = {
    'AK': 0,
    'BCC': 1,
    'BKL': 2,
    'DF': 3,
    'MEL': 4,
    'NV': 5,
    'SCC': 6,
    'VASC': 7
}

# Convert class labels to numerical values using the mapping
y_train_encoded = np.array([class_mapping[label] for label in y_train])
y_test_encoded = np.array([class_mapping[label] for label in y_test])

num_classes = len(class_mapping)
# Convert encoded class labels to one-hot encoded vectors
y_train_categorical = to_categorical(y_train_encoded, num_classes)
y_test_categorical = to_categorical(y_test_encoded, num_classes)
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3))) 
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(8, activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer ='adam',
              metrics = ['accuracy'])
```

```python
# Assuming you have already one-hot encoded your labels as y_train_categorical and y_test_categorical
# Convert the data type of the labels to float32
y_train_categorical = y_train_categorical.astype('float32')
y_test_categorical = y_test_categorical.astype('float32')

# Train the model with a separate validation dataset
history = model.fit(X_train, y_train_categorical, 
                    batch_size=32, 
                    epochs=30, 
                    verbose=1, 
                    validation_split=0.2)
```

```python
plt.rcParams["figure.figsize"] = (5,7)
plt.plot(history.history['loss'], label='Train_loss')
plt.plot(history.history['val_loss'], label='Val_loss')
plt.ylim([0, 3])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import shutil


# Define the base directory
base_dir = 'skin_lesion_images'

# Create train and test directories
train_dir = os.path.join(base_dir, 'train')
test_dir = os.path.join(base_dir, 'test')

# Create directories for each class in train and test directories
classes = ['AK', 'BCC', 'BKL', 'DF', 'MEL', 'NV', 'SCC', 'VASC']
for class_name in classes:
    os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(test_dir, class_name), exist_ok=True)

# Iterate over the original images and move them to the appropriate directories
for class_name in os.listdir(base_dir):
    class_dir = os.path.join(base_dir, class_name)
    if os.path.isdir(class_dir):
        image_files = [f for f in os.listdir(class_dir) if f.endswith('.jpg')]
        for i, image_file in enumerate(image_files):
            src = os.path.join(class_dir, image_file)
            dest_dir = train_dir if i % 5 != 0 else test_dir  # 80% train, 20% test
            dest = os.path.join(dest_dir, class_name, image_file)
            shutil.copy(src, dest)

print("Dataset organized successfully.")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

base_dir = 'skin_lesion_images'
train_dir = os.path.join(base_dir, 'train')
test_dir = os.path.join(base_dir, 'test')

# Define image data generators for training and validation
train_datagen = ImageDataGenerator(rescale=1./255)  # Normalize pixel values to [0, 1]
test_datagen = ImageDataGenerator(rescale=1./255)

# Define batch size
batch_size = 32

# Create data generators
train_generator = train_datagen.flow_from_directory(
    train_dir,  # Directory containing training images
    target_size=(256, 256),  # Resize images to 256x256
    batch_size=batch_size,
    class_mode='categorical'  # Use categorical labels
)

test_generator = test_datagen.flow_from_directory(
    test_dir,  # Directory containing test images
    target_size=(256, 256),  # Resize images to 256x256
    batch_size=batch_size,
    class_mode='categorical'  # Use categorical labels
)

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Define the DNN model
model = Sequential([
    Flatten(input_shape=(256, 256, 3)),  # Flatten the input image to a 1D array
    Dense(64, activation='relu'),  # Add a fully connected layer with 64 neurons and ReLU activation
    Dense(32, activation='relu'),  # Add another fully connected layer with 32 neurons and ReLU activation
    Dense(num_classes, activation='softmax')  # Add the output layer with softmax activation
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model using the generators
history = model.fit(train_generator, epochs=30, steps_per_epoch=len(X_train) // batch_size,
                    validation_data=test_generator, validation_steps=len(X_test) // batch_size)

```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
